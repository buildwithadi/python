```python
import pandas as pd
```

```python
# Series

data = [1,2,3,4,5,6]
series = pd.Series(data)
print("Series:\n",series)
print(type(series))
```

```python
data = {'a':1,'b':2,'c':3}
series_dict = pd.Series(data)
print(series_dict)
```

```python
data = [10,20,30]
index = ['a','b','c']
pd.Series(data, index = index)
```

```python
# Dataframe

data = {
    'Name':['Adi','Krish','John'],
    'Age':[25,30,45],
    'City':['Banglore','New York','Florida']
}

df = pd.DataFrame(data)
print(df)
print(type(df))
```

```python
data = [
    {'Name':'Adi','Age':20,'City':'Banglore'},
    {'Name':'Krish','Age':30,'City':'Banglore'},
    {'Name':'Sam','Age':28,'City':'Delhi'},
]

df = pd.DataFrame(data)
print(df)
```

```python
data = [
    {'Name':'Adi','Age':20,'City':'Banglore'},
    {'Name':'Krish','Age':30,'City':'Banglore'},
    {'Name':'Sam','Age':28,'City':'Delhi'},
]

df = pd.DataFrame(data)
print(df)
```
