```python
import matplotlib.pyplot as plt
```

```python
x = [1,2,3,4,5]
y = [1,4,9,16,25]

# Line Plot
plt.plot(x,y)
plt.xlabel("X axis")
plt.ylabel('Y axis')
plt.title("Line Plot")
plt.show()
```

```python
x = [1,2,3,4,5]
y = [1,4,9,16,25]

# Customized Line Plot
plt.plot(x,y, color='red', linestyle='--', marker='o', linewidth=3, markersize=9)
plt.grid(True)
plt.show()
```

```python
## Multiple Plots

x = [1,2,3,4,5]
y1 = [1, 4, 9 ,16, 25]
y2 = [11,22,33,44,55]

plt.figure(figsize=(9,5))

plt.subplot(2,2,1)  # 1st block of 2 by 2 grid
plt.plot(x, y1, color='green')
plt.title("Plot 1")

plt.subplot(2,2,2)  # 2st block of 2 by 2 grid
plt.plot(y1, x, color='red')
plt.title("Plot 2")

plt.subplot(2,2,3)  # 3st block of 2 by 2 grid
plt.plot(x, y2, color='blue')
plt.title("Plot 3")

plt.subplot(2,2,4)  # 4st block of 2 by 2 grid
plt.plot(x, y2, color='cyan')
plt.title("Plot 4")

plt.show()
```

```python
# Bar Plot
categories = ['A', 'B', 'C', 'D', 'E']
values = [23,53,12,34,23]

plt.bar(categories, values, color='purple')
plt.xlabel('Categories')
plt.ylabel('Bar Plot')
plt.show()
```

```python
## Histogram

data = [1,2,1,4,5,6,6,7,5,5,3,7,2,4,7,3,4]

plt.hist(data, bins=5, color='cyan', edgecolor='black')
```

```python
## Scatter Plot
x = [1,2,3,4,5]
y = [2,3,4,5,6]

plt.scatter(x, y, color='blue', marker='o')
```

```python
# Pie Chart

labels = ['A','B', 'C', 'D']
sizes = [30, 20, 40, 10]
colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue']
explode = (0.2,0,0,0)

plt.pie(sizes, explode=explode, labels=labels, colors=colors, autopct="%1.1f%%")
```

```python
# Sales Data Visualization

import pandas as pd
sales_data_df = pd.read_csv('sales_data.csv')
sales_data_df.head()
```

```python
sales_data_df.info()
```

```python
# Plot total sales by products
total_sales_by_product = sales_data_df.groupby(['Product Category'])['Total Revenue'].sum()

total_sales_by_product
```

```python
total_sales_by_product.plot(kind='bar', color='cyan')
```

```python
date = list(sales_data_df['Date'])
months = list()

for i in date:
    months.append(i.split('-')[1])

sales_data_df['Months'] = months
```

```python
sales_trend=sales_data_df.groupby('Months')['Total Revenue'].sum().reset_index()
plt.plot(sales_trend['Months'],sales_trend['Total Revenue'])
```
